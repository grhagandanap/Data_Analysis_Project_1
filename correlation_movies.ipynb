```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

plt.style.use('ggplot')
df = pd.read_csv('D:\Programmingkeun\SQL\Data Analyst Project 1\movies.csv')
df.head()
df.info()

# Check the missing data
for col in df.columns:
    data_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,data_missing))
```

```python
#checking data type
df.dtypes
```

```python
#change data type

df['gross'] = df['gross'].astype('Int64')

df['budget'] = df['budget'].astype("Int64")
```

```python
#splitting year

df['year_correct'] = df["released"].str.split(" ",n=3,expand = True)[2]
df['year_correct']
```

```python
#sorting data by 'gross'

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duolicates

df.drop_duplicates()
```

```python
# finding correlation

# scatter plot with budget vs gross
df['gross'] = df['gross'].astype('float64')
df['budget'] = df['budget'].astype('float64')
plt.figure(figsize=(15,8))
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.legend()
plt.show()
```

```python
# plot budget vs gross in sns

plt.figure(figsize=(15,8))
sns.regplot(x='budget', 
    y='gross', 
    data=df, 
    scatter_kws={"color":"orange"}, 
    line_kws={"color":"blue"})
plt.title('Budget vs Gross Earnings')
```

```python
#checking correlation for each variable

df.corr(method='kendall') #pearson, kendall, spearman
```

```python
#correlation viz

correlation_matrix = df.corr(method='pearson')

plt.figure(figsize=(15,8))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Heatmap for Variables')
plt.xlabel('Variables')
plt.ylabel('Variables')
```

```python
#numerizing all variables

df_numerized = df

for col in df_numerized:
    if(df_numerized[col].dtype == 'object'):
        df_numerized[col] = df_numerized[col].astype('category')
        df_numerized[col] = df_numerized[col].cat.codes

df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

plt.figure(figsize=(15,8))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Heatmap for Variables')
plt.xlabel('Variables')
plt.ylabel('Variables')
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.sort_values()

corr_pairs[(corr_pairs>0.5)].sort_values()
```
